## merge

```python
df1 = pd.DataFrame({
    "고객번호" : [1001, 1002, 1003, 1004],
    "이름" : ["둘리", "도우너", "또치", "고길동"]
}, columns=["고객번호", "이름"])
df1.head()
```

```python
df2 = pd.DataFrame({
    "고객번호" : [1001, 1002, 1003, 1004],
    "금액" : [1000, 30000, 20000, 15000]
}, columns=["고객번호", "금액"])
df2.head()
```

```python
df = pd.merge(df1, df2)
df
```

```python
df1 = pd.DataFrame({
    "고객번호" : [1001, 1002, 1003, 1004, 1005],
    "이름" : ["둘리", "도우너", "또치", "고길동", "홍영희"]
}, columns=["고객번호", "이름"])
df1.head()

df2 = pd.DataFrame({
    "고객번호" : [1001, 1002, 1003, 1004],
    "금액" : [1000, 30000, 20000, 15000]
}, columns=["고객번호", "금액"])
df2.head()

df = pd.merge(df1, df2)
df
# inner join이라서 1005번은 나오지 않음
```

```python
df = pd.merge(df1, df2, how="outer")
df
# how 파라미터로 join 설정 가능
# inner, outer, left, right
# 참조 : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html
```

```python
df = pd.merge(df1, df2, how="left")
df
```

```python
df = pd.merge(df1, df2, how="right")
df
```

```python
pd.merge(df1, df2, on="고객번호")
# 무엇을 공통으로 묶을 것이냐
```

## 인덱스 기준

```python
df1 = pd.DataFrame({
    "도시" : ["서울", "서울", "서울", "서울", "부산", "부산"],
    "년도" : [2000, 2005, 2010, 2009, 2005, 2006],
    "인구" : [9834949, 9934324, 1000034, 1000034, 5000000, 6000000]
})
df1.head()
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape(6,2),
    index=[["부산", "부산", "서울", "서울", "서울", "서울"],
           [2005, 2006, 2000, 2005, 2003, 2010]],
    columns=[["데이터1", "데이터2"]]
)
df2.head()
```

```python
df_r = pd.merge(df1, df2, left_on=["도시", "년도"], right_index=True)
df_r
```

```python
# how=outer : 겹치는 column 이 없어도 일단 합쳐줌
```
