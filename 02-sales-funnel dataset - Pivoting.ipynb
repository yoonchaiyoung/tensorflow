```python
import pandas as pd
df = pd.read_excel("02-sales-funnel.xlsx")
df.head(3)
```

```python
pd.pivot_table(df, index=["Product"])
# Product를 기준으로 sorting 함
# 숫자로 된 column 만
```

```python
pd.pivot_table(df, index=["Name", "Rep", "Manager"])
```

```python
import numpy as np
pd.pivot_table(df, index=["Rep", "Manager"], 
               values=["Price"],
              aggfunc=np.sum)
```

```python
import numpy as np
pd.pivot_table(df, index=["Rep", "Manager"], 
               values=["Price"],
              aggfunc=[np.sum, np.mean])
```
